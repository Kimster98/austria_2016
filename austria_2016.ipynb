```python
#Sum of first round votes per candidate
sum_hofer_1 = df['Hofer'].sum()
sum_vdb_1 = df['Van der Bellen'].sum()
sum_griss_1 = df['Griss'].sum()
sum_hundstorfer_1 = df['Hundstorfer'].sum()
sum_khol_1 = df['Khol'].sum()
sum_lungner_1 = df['Lungner'].sum()
```

```python
first_round = [['Hofer', df['Hofer'].sum(), round(100*(df['Hofer'].sum()/df['Sum of Votes'].sum()),2)],
               ['Van der Bellen', df['Van der Bellen'].sum(), round(100*(df['Van der Bellen'].sum()/df['Sum of Votes'].sum()),2)],
               ['Griss', df['Griss'].sum(), round(100*(df['Griss'].sum()/df['Sum of Votes'].sum()),2)],
               ['Hundstorfer', df['Hundstorfer'].sum(), round(100*(df['Hundstorfer'].sum()/df['Sum of Votes'].sum()),2)],
               ['Khol', df['Khol'].sum(), round(100*(df['Khol'].sum()/df['Sum of Votes'].sum()),2)],
               ['Lungner', df['Lungner'].sum(), round(100*(df['Lungner'].sum()/df['Sum of Votes'].sum()),2)]]

#dataframe with results of the first round
df3 = pd.DataFrame(first_round, columns=['Name', 'Votes', 'Percent'])
df3

```

```python
# Visualizing the results of the first round
# No candidate finished above 50%
# A second round was held between Hofer and Van der Bellen

color_scale = alt.Scale(
    domain=df3['Name'].unique(),
    range=['blue', 'yellow', 'pink', 'red', 'black', 'purple']
)

chart = alt.Chart(df3).mark_bar().encode(
    x=alt.X('Percent:Q', axis=alt.Axis(title='Percent of valid votes'), scale=alt.Scale(domain=[0, 60])),
    y=alt.Y('Name:N', sort='-x', axis=alt.Axis(title='Candidates'), band=0.9),
    color=alt.Color('Name:N', scale=color_scale),
    tooltip=[
        alt.Tooltip('Name:N', title='Candidate'),
        alt.Tooltip('Votes:Q', title='Votes', format=','),
        alt.Tooltip('Percent:Q', title='Percent', format=','),
    ]
).properties(
    title='First Round Results'
)

threshold = alt.Chart(pd.DataFrame({'threshold': [50]})).mark_rule(color='darkred').encode(
    x='threshold:Q'
)

chart + threshold
```

```python
# Dataframe with result of the second round

second_round = [['Van der Bellen', df['Van der Bellen 2nd round'].sum(), round(100*(df['Van der Bellen 2nd round'].sum()/df['Sum of Votes 2nd round'].sum()),2)],
                ['Hofer', df['Hofer 2nd round'].sum(), round(100*(df['Hofer 2nd round'].sum()/df['Sum of Votes 2nd round'].sum()),2)]]

df4 = pd.DataFrame(second_round, columns=['Name', 'Votes', 'Percent'])
df4
```

```python
# Visualizing the results of the second round

color_scale_2 = alt.Scale(
    domain=df4['Name'].unique(),
    range=['yellow', 'blue']
)


chart = alt.Chart(df4).mark_bar().encode(
    x=alt.X('Percent:Q', axis=alt.Axis(title='Percent of valid votes'), scale=alt.Scale(domain=[0, 60])),
    y=alt.Y('Name:N', sort='-x', axis=alt.Axis(title='Candidates'), band=0.9),
    color=alt.Color('Name:N', scale=color_scale_2),
    tooltip=[
        alt.Tooltip('Name:N', title='Candidate'),
        alt.Tooltip('Votes:Q', title='Votes', format=','),
        alt.Tooltip('Percent:Q', title='Percent', format=',')
    ]
).properties(
    title='Second Round Results'
)

chart
```

```python
# Difference between Van der Bellen and Hofer in the federal states

df2["Difference"] = df2["Van der Bellen 2nd round %"] - df2["Hofer 2nd round %"]

chart = alt.Chart(df2).mark_bar().encode(
    x=alt.X("Difference", title="Difference (%)"),
    y=alt.Y("Bundesland", sort=alt.EncodingSortField(field="Difference", order='ascending'), band=0.9),
    color=alt.condition(
        alt.datum.Difference > 0,
        alt.value("yellow"),
        alt.value("blue")
    ),
    tooltip=[alt.Tooltip("Bundesland", title="Bundesland"), alt.Tooltip("Difference", title="Difference (%)")]
).properties(
    title="Difference between Van der Bellen and Hofer in the second round"
)

chart
```

I read about this election in a book, and it was mentioned, that Van der Bellen could win, because the majority of the supporters of the √ñVP candidate (Khol) also supported him in the second round.

Initially, I wanted to figure out whether this claim is true or not, using the election data.

What can be said for sure, that in every district, Van der Bellen increased his vote share in the second round with more than half of the sum of the vote shares of the candidates, who didn't make it to the second round.

```python
# In the last two figures, the sum of the vote shares of the candidates who did not make it out of the first round is on the x-axis (this is the "Other %")
# On the y-axis is the difference between the first round result of Van der Bellen (first figure) and Hofer (second figure)
# Each point represents a district
# Yellow point means Van der Bellen got more votes in the district, blue point means Hofer got more votes in the district
# Unsurprisingly, the majority of voters for the candidates who were eliminated in the first round voted for Van der Bellen in the second round
# There is no constituency where Van der Bellen's share of the vote has not increased by at least half of the share of the vote of the candidates who were eliminated
# All points in Van der Bellen's graph are above the red line indicating this, while all points in Hofer's graph are below the red line



df['Van der Bellen gain %'] = df['Van der Bellen 2nd round %'] - df['Van der Bellen %']
df['Hofer gain %'] = df['Hofer 2nd round %'] - df['Hofer %']
df['Other %'] = df['Griss %'] + df['Hundstorfer %'] + df['Khol %'] + df['Lungner %']
df['Winner'] = df['Van der Bellen 2nd round %'] > 50

scatter = alt.Chart(df).mark_point(filled=True,size=100).encode(
    x=alt.X('Other %', scale=alt.Scale(domain=(32, 60))),
    y=alt.Y('Van der Bellen gain %', scale=alt.Scale(domain=(20, 42))),
    color=alt.condition(
        alt.datum.Winner,
        alt.value('yellow'),
        alt.value('blue')),
    tooltip=[alt.Tooltip('Bundesland:N', title='Bundesland'),
             alt.Tooltip('Bezirk:N', title='Bezirk'),
             alt.Tooltip('Van der Bellen 2nd round %', title = 'Van der Bellen 2nd round %'),
             alt.Tooltip('Van der Bellen gain %', title='Van der Bellen gain %'),
             alt.Tooltip('Other %', title='Other %')]
).properties(
    title='Increase of Van der Bellens vote shares in the districts',
    width=500,
    height=500
)


data = [[40, 20], [60, 30]]
df7 = pd.DataFrame(data, columns=['Other %', 'Van der Bellen gain %'])

red_line = alt.Chart(df7).mark_line(color='red').encode(
    alt.X('Other %'),
    alt.Y('Van der Bellen gain %')
)
scatter + red_line

```

Of course it doesn't mean for sure that the majority of Griss, Hundstorfer, Khol or Lungner voters supported Van der Bellen in the second round, but it makes it very likely.

I tried to figure out how Griss, Khol and Hundstorfer voters behaved in the second round. (I ignored Lungner voters because he got only around 2% of the votes)

Before that, I already noticed that there are some interesting districts, for example Wels (Stadt). Hofer performed strongly there in the first round, getting 39% of the votes, however he could only improve 4%-points on that result in the second round.


Especially when talking about voters of Griss - in some Viennese districts Hofer could barely improve his share of votes for the second round (gained less then 5%-points in several districts), while Griss got around 20% of the votes in these districts.




```python
hofer_national_gain = df4['Percent'][1]-df3['Percent'][0]
hofer_national_gain
```

```python
df_griss_hofergain = pd.DataFrame()
df_griss_hofergain['Bundesland'] = df['Bundesland']
df_griss_hofergain['Bezirk'] = df['Bezirk']
df_griss_hofergain['Griss %'] = df['Griss %']
df_griss_hofergain['Hofer gain %'] = df['Hofer gain %']
df_griss_hofergain['Griss % / Hofer gain %'] = round(df_griss_hofergain['Griss %']/df_griss_hofergain['Hofer gain %'], 3)
df_griss_hofergain = df_griss_hofergain.sort_values(by = 'Griss % / Hofer gain %', ascending = False)
df_griss_hofergain
```

```python
print(len(df_griss_hofergain[df_griss_hofergain['Griss % / Hofer gain %']>2]))
```

```python
print(len(df_griss_hofergain[df_griss_hofergain['Griss % / Hofer gain %']>1]))
```

```python
print(len(df_griss_hofergain[df_griss_hofergain['Griss % / Hofer gain %']>df3['Percent'][2]/hofer_national_gain]))
```

It is visible that only in districts in Burgenland (the weakest district of Griss) and Nieder√∂sterreich could Hofer gain more percent points than the first round result of Griss. At the same time, in most of the districts (exactly 99 districts) are below the red line, so Hofer could gain less percent points than the first round result of Griss. (It could also be because of the fact that Griss got the most of the votes of the "fallen-out" candidates)

```python
scatter_plot = alt.Chart(df_griss_hofergain).mark_circle(filled=True, size=75).encode(
    x=alt.X('Griss %'),
    y=alt.Y('Hofer gain %'),
    tooltip=['Bundesland', 'Bezirk', 'Griss %', 'Hofer gain %', 'Griss % / Hofer gain %'],
    color='Bundesland:N'
).properties(
    width=500,
    height=400
)


data = [[0, 0], [30, 30]]
df10 = pd.DataFrame(data, columns=['Griss %', 'Hofer gain %'])

red_line = alt.Chart(df10).mark_line(color='red').encode(
    alt.X('Griss %'),
    alt.Y('Hofer gain %')
).properties(
    title='Griss versus Hofer gain in the districts'
)

# Combine scatter plot and red line
chart = scatter_plot + red_line

# Show the chart
chart
```
